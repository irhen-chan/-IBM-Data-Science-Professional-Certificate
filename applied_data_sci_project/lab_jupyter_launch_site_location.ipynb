```python
import folium
from folium.features import DivIcon


# 1) Initialize the map
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# 2) Loop through each row in launch_sites DataFrame
for index, row in launch_sites_df.iterrows():
    # Extract coordinates and launch site name
    site_lat = row['Lat']
    site_lon = row['Long']
    site_name = row['Launch Site']
    
    # 3) Add a Circle at each launch site
    folium.Circle(
        location=[site_lat, site_lon],
        radius=1000,              # radius in meters
        color='#d35400',          # circle line color
        fill=True,
        fill_color='#FFA07A'      # circle fill color
    ).add_child(
        folium.Popup(site_name)   # popup text with site name
    ).add_to(site_map)
    
    # 4) Add a Marker (using DivIcon for text label)
    folium.map.Marker(
        [site_lat, site_lon],
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % site_name
        )
    ).add_to(site_map)

# 5) Display the map
site_map
```


#FINDINGS:

Proximity to the Equator

The Equator sits at 0° latitude, whereas these U.S. launch sites (in Florida and California) are roughly between 28° N and 35° N. This means they are not particularly close to the Equator. While launch sites closer to the equator can take advantage of Earth’s rotational speed, Florida and southern California provide a balance of other logistical considerations (e.g., weather, large over-water flight corridors, existing space infrastructure).
Proximity to the Coast

All of the marked launch sites (CCAFS SLC-40, KSC LC-39A in Florida, and VAFB SLC-4E in California) are indeed located near the coast. This is intentional:
Launches typically aim over open water for safety in case of anomalies.
Coastal locations allow rocket stages to land or drop over the ocean, minimizing risk to populated areas.

```python
import folium
from folium.plugins import MarkerCluster

# 1) Suppose `spacex_df` has columns: ['Lat', 'Long', 'class', 'LaunchSite']
#    Make a new column 'marker_color' based on 'class'.
spacex_df['marker_color'] = spacex_df['class'].apply(
    lambda c: 'green' if c == 1 else 'red'
)

# 2) Initialize the MarkerCluster
marker_cluster = MarkerCluster()

# 3) Loop and add Markers
for idx, row in spacex_df.iterrows():
    coord = [row['Lat'], row['Long']]
    m_color = row['marker_color']   # Make sure this matches exactly
    popup_text = f"Site: {row['Launch Site']}, Class: {row['class']}"
    
    marker = folium.Marker(
        location=coord,
        icon=folium.Icon(color=m_color, icon='info-sign'),
        popup=popup_text
    )
    marker_cluster.add_child(marker)

# 4) Create a Folium map, add cluster, and display
site_map = folium.Map(location=[29.559684, -95.083097], zoom_start=5)
site_map.add_child(marker_cluster)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
from math import sin, cos, sqrt, atan2, radians
import folium
from folium.features import DivIcon

# 1) Define a helper function to calculate distance between two lat/lon pairs
def calculate_distance(lat1, lon1, lat2, lon2):
    """
    Returns the great-circle distance in KM between two points 
    specified by lat/long using the haversine formula.
    """
    # approximate radius of Earth in km
    R = 6373.0

    # Convert lat/lon from degrees to radians
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance

# 2) Suppose you have a launch site coordinate, e.g., 
launch_site_lat = 28.563197
launch_site_lon = -80.576820

# And you pick a coastline point (via MousePosition or manual)
# For example:
coastline_lat = 28.56367
coastline_lon = -80.57163

# 3) Calculate the distance
distance_coastline = calculate_distance(
    launch_site_lat, launch_site_lon, 
    coastline_lat, coastline_lon
)

print("Distance from launch site to coastline: {:.2f} km".format(distance_coastline))

# 4) Create a Folium map, centered roughly at the launch site
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=14)

# 5) Add a marker for the launch site (optional)
folium.Marker(
    [launch_site_lat, launch_site_lon],
    icon=folium.Icon(color='blue', icon='info-sign'),
    popup=f"Launch Site ({launch_site_lat}, {launch_site_lon})"
).add_to(site_map)

# 6) Create and add a Marker for the chosen coastline point
#    Show the distance in the label (using DivIcon)
distance_marker = folium.Marker(
    [coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html=(
            '<div style="font-size: 12; color:#d35400;">'
            f'<b>Coastline Point</b><br>'
            f'{distance_coastline:.2f} km</div>'
        )
    )
)
distance_marker.add_to(site_map)

# 7) Draw a PolyLine between the coastline and the launch site
points = [(launch_site_lat, launch_site_lon),
          (coastline_lat, coastline_lon)]
line = folium.PolyLine(locations=points, weight=2)
site_map.add_child(line)

# 8) Display the map
site_map
```
